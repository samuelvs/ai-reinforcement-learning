```python
# Visualizando estados:
import time

env = gym_super_mario_bros.make('SuperMarioBros-1-1-v0')
env = JoypadSpace(env, RIGHT_ONLY)
obs = env.reset()

for i in range(500):
    env.render()
    time.sleep(0.02)
    acao = 1
    obs, recompensa, terminou, info = env.step(acao)
    print(recompensa)
    if terminou:
        print("Episódio finalizado depois de {} steps".format(i+1))
        break
env.close()
```

```python
import gym_super_mario_bros
from nes_py.wrappers import JoypadSpace
from gym_super_mario_bros.actions import RIGHT_ONLY
import time

from OpenAIWrappers import env_wrapper
from stable_baselines3.common.vec_env import SubprocVecEnv
from stable_baselines3 import PPO
from stable_baselines3.common.callbacks import CheckpointCallback

def build_env(name_env):
    def _init():
        env = gym_super_mario_bros.make(name_env)
        env = JoypadSpace(env, RIGHT_ONLY)
        env = env_wrapper(env)
        return env
    return _init


threads_cpu = 8
env = SubprocVecEnv([def build_env(name_env='SuperMarioBros-1-1-v0') for i in range(threads_cpu)])

checkpoint_callback = CheckpointCallback(save_freq=6250, save_path='C:/Users/samue/Documents/modelos_salvos/', name_prefix='n1024b64l3')

#model = PPO(policy='CnnPolicy', env=env, gamma=0.99, n_steps=1024, batch_size=64, learning_rate=0.0003, vf_coef=0.5, ent_coef=0.01, verbose=1)
model = PPO.load('C:/Users/samue/Documents/modelos_salvos/n1024b64l3_1200000_steps.zip', env=env, verbose=1)

start = time.time()
model.learn(total_timesteps=10000000, log_interval=10, callback=checkpoint_callback)

end = time.time()
print('Tempo total do treinamento em horas:', (end-start)/3600)
```
